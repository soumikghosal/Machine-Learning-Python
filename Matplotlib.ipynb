```python
import numpy as np
t = np.linspace(-np.pi,np.pi, 361)
sin=np.sin(t)
cos=np.cos(t)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.plot(t,sin)
plt.plot(t,cos)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle=":")   # line type=':' '-.' 'steps' #linestyle=ls #linewidth=lw
plt.plot(X, S, 'r--', linewidth=2.5, marker=3)

plt.xlim(-5,5)
plt.ylim(-1,1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])     #the values we need to see on the axis
plt.yticks([-1, 0, +1])

plt.show()
```

```python
plt.plot(t,sin,'rs',t,cos,'b^')  #rs- red square; b^- blue triangle
plt.xlabel("Angles", fontsize=15)

plt.legend( ['Sine Curve','Cosine Curve'],loc='upper left',fontsize=11)

ax = plt.gca()  # gca stands for 'get current axis'

ax.spines['right'].set_color('none')   #'none' - invisible right spine
ax.spines['top'].set_color('none')   #'none' - invisible right spine

ax.spines['bottom'].set_position(('data',0))  #position the x axis
ax.spines['left'].set_position(('data',0))    #position the y axis

ax.xaxis.set_ticks_position('bottom')  #position of x axis ticks
ax.yaxis.set_ticks_position('left')   #position of y axis ticks

plt.show()
```
